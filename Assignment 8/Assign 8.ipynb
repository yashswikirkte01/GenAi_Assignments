```python
import torch
import torchaudio
from torchaudio.pipelines import TACOTRON2_WAVERNN_PHONE_LJSPEECH

# Load model bundle
bundle = TACOTRON2_WAVERNN_PHONE_LJSPEECH

processor = bundle.get_text_processor()
tacotron2 = bundle.get_tacotron2()
vocoder = bundle.get_vocoder()

text = "Hello, welcome to deep learning text to speech synthesis."

# Convert text to phoneme tokens
processed, lengths = processor(text)

# Generate spectrogram
with torch.inference_mode():
    spec, spec_lengths, _ = tacotron2.infer(processed, lengths)

# Convert spectrogram to waveform
waveforms, lengths = vocoder(spec, spec_lengths)

# Save audio
torchaudio.save("speech.wav", waveforms[0:1], 22050)

print("Speech saved as speech.wav")
```
